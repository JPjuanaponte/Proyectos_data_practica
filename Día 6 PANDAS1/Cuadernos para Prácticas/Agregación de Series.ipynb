```python
import pandas as pd
ventas = [120, 150, 90, 200, 210, 130, 160]
indice = ['Lunes', 'Martes', 'Miercoles', 'Jueves', 'Viernes', 'sabado', 'Domingo']

serie = pd.Series(ventas, index= indice)

suma_total_ventas = serie.sum()
print(f"Suma total: {suma_total_ventas}")
dia_mayores_ventas = serie.idxmax()
print(f"Día de mayores ventas: {dia_mayores_ventas}")
promedio_ventas = serie.mean()
print(promedio_ventas)


```
