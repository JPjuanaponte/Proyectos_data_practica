# LIMPIEZA DE DATOS

```python
# llenar valores nulos en 0 , porque no se gano medalla y no cambia la informacipond elos datos 

df.fillna(0, inplace=True) # inplace modifica el dataframe , no copia.
df.info()
```

```python
# corregir tipo de datos

df['Oro']= df['Oro'].astype(int)
df['Plata']= df['Plata'].astype(int)
df['Bronce']= df['Bronce'].astype(int)
```

```python
df.head(10)
```

# ANÁLISIS MEDALLAS DE ORO

```python
df_oro = df.sort_values(by='Oro', ascending=False).head(3) # ASCENDING DE MAYOR A MENOR EN LA COLUMAN QUE YO LE INDIQUE
df_oro
```

# ANÁLISIS GANARON MAS DE 10 MEDALLAS

```python
condicion = df[df['Total']>10 ]
print(condicion)
```

```python
condicion = df[df['Total']>10 ].sort_values(by='Total',ascending=False )
print(condicion)
```
